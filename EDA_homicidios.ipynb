# Crear una función para reorganizar el texto de la calle

Por motivos esteticos personales , deseamos pasar al primer puesto de la cadena de texto , todos los valores de la columna "Calle" que contengan la abrevacion "AV".

```python


def reorganizar_calle(texto):
    partes = texto.split()  # Dividir la cadena en partes en función del espacio
    if len(partes) > 1:
        nueva_calle = partes[-1] + " " + " ".join(partes[:-1])  # Reorganizar las partes
        return nueva_calle
    else:
        return texto

# Aplicar la función a la columna 'calle' del DataFrame
dfh1['Calle'] = dfh1['Calle'].apply(reorganizar_calle)
```

```python
# Función para reorganizar el texto de la calle
def reorganizar_calle(texto):
    partes = texto.split(",")  # Dividir la cadena en partes en función de la coma
    if len(partes) > 1:
        if "AV." in partes[0]:  # Verificar si "AV." está presente en la primera parte
            nueva_calle = "AV. " + partes[0].replace("AV.", "").strip() + " " + partes[1].strip()  # Reorganizar las partes
            return nueva_calle
        else:
            nueva_calle = "AV. " + partes[1].strip() + " " + partes[0].strip()  # Reorganizar las partes
            return nueva_calle
    else:
        return texto

# Aplicar la función a la columna 'calle' del DataFrame
dfh1['Calle'] = dfh1['Calle'].apply(reorganizar_calle)

```

# Calcular la frecuencia de accidentes para cada hora del día

Podemos observa mediante nuestro grafico que la hora con mayor frecuencia en los años estudiados son las 7am , por otro lado tenemos dos valores de igual valor en la menor frecuencia presente en la mañana  , nos referimos a las 2am y 13pm.

```python

frecuencia_por_hora = dfh1['HH'].value_counts().sort_index()

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
frecuencia_por_hora.plot(kind='bar', color='yellow')
plt.xlabel('Hora del Día')
plt.ylabel('Número de Accidentes')
plt.title('Número de Accidentes por Hora del Día')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Calcular la frecuencia de accidentes para cada tipo de calle
frecuencia_por_tipo_calle = dfh1['TIPO_DE_CALLE'].value_counts()

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
frecuencia_por_tipo_calle.plot(kind='bar', color='skyblue')
plt.xlabel('Tipo de Calle')
plt.ylabel('Número de Accidentes')
plt.title('Número de Accidentes por Tipo de Calle')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Agrupar los datos por año y tipo de víctima, y calcular la frecuencia de accidentes
frecuencia_por_anio_victima = dfh1.groupby(['AAAA', 'VICTIMA']).size().reset_index(name='Frecuencia')

# Encontrar el tipo de víctima más frecuente para cada año
victima_mas_frecuente_por_anio = frecuencia_por_anio_victima.loc[frecuencia_por_anio_victima.groupby('AAAA')['Frecuencia'].idxmax()]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(victima_mas_frecuente_por_anio['AAAA'], victima_mas_frecuente_por_anio['Frecuencia'], color='skyblue')
plt.xlabel('Año')
plt.ylabel('Frecuencia')
plt.title('Tipo de Víctima más Frecuente por Año')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python


# Filtrar los datos para seleccionar solo los accidentes que ocurrieron en el año 2018
accidentes_2018 = dfh1[dfh1['AAAA'] == 2018]

# Agrupar los datos por tipo de víctima y tipo de calle, y calcular la frecuencia de accidentes para cada combinación
frecuencia_por_victima_calle_2018 = accidentes_2018.groupby(['VICTIMA', 'TIPO_DE_CALLE']).size().reset_index(name='Frecuencia')

# Encontrar la víctima más frecuente en el año 2018
victima_mas_frecuente_2018 = frecuencia_por_victima_calle_2018.loc[frecuencia_por_victima_calle_2018['Frecuencia'].idxmax()]

# Imprimir la víctima más frecuente, el tipo de calle más común y la frecuencia de accidentes correspondiente al año 2018
print("La víctima más frecuente en el año 2018 fue:", victima_mas_frecuente_2018['VICTIMA'])
print("Tipo de calle más común en el año 2018:", victima_mas_frecuente_2018['TIPO_DE_CALLE'])
print("Frecuencia de accidentes:", victima_mas_frecuente_2018['Frecuencia'])

# Filtrar los datos para la víctima más frecuente en 2018
victima_mas_frecuente_2018_data = frecuencia_por_victima_calle_2018[frecuencia_por_victima_calle_2018['VICTIMA'] == victima_mas_frecuente_2018['VICTIMA']]

# Crear el gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(victima_mas_frecuente_2018_data['Frecuencia'], labels=victima_mas_frecuente_2018_data['TIPO_DE_CALLE'], autopct='%1.1f%%', startangle=140)
plt.title('Porcentaje de Frecuencia de Accidentes por Tipo de Calle para la Víctima más Frecuente en 2018')
plt.axis('equal')
plt.show()
```

```python
# Obtener el tipo de calle más común
tipo_calle_mas_comun = victima_mas_frecuente_2018_data.loc[victima_mas_frecuente_2018_data['Frecuencia'].idxmax()]['TIPO_DE_CALLE']

# Filtrar los datos para el tipo de calle más común
datos_tipo_calle_mas_comun = dfh1[dfh1['TIPO_DE_CALLE'] == tipo_calle_mas_comun]

# Calcular la frecuencia de tipo de víctima
frecuencia_tipo_victima = datos_tipo_calle_mas_comun['VICTIMA'].value_counts()

# Calcular la frecuencia de tipo de acusado
frecuencia_tipo_acusado = datos_tipo_calle_mas_comun['ACUSADO'].value_counts()

# Imprimir los resultados
print("Tipo de calle más común en 2018:", tipo_calle_mas_comun)
print("Tipo de víctima más presente en este tipo de calle:", frecuencia_tipo_victima.idxmax())
print("Tipo de acusado más común en este tipo de calle:", frecuencia_tipo_acusado.idxmax())
```

# Calcular el número total de víctimas en motocicleta por año


```python


# Convertir la columna de fecha a tipo datetime
dfh1['FECHA'] = pd.to_datetime(dfh1['FECHA'])

# Crear una nueva columna para el año
dfh1['AÑO'] = dfh1['FECHA'].dt.year

# Filtrar los datos solo para las víctimas en motocicleta
victimas_moto = dfh1[dfh1['VICTIMA'] == 'MOTO']

# Calcular el número total de víctimas en motocicleta por año
victimas_moto_por_año = victimas_moto.groupby('AÑO').size()

# Crear el gráfico de tendencia de víctimas en motocicleta a lo largo de los años
plt.figure(figsize=(10, 6))
plt.plot(victimas_moto_por_año.index, victimas_moto_por_año.values, marker='o')
plt.title('Tendencia de víctimas en motocicleta a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Total de víctimas en motocicleta')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

```

Podemos obsevar como la tendencia a traves de los años va disminuyendo hasta el año 2020 donde tiene su punto mas bajo , al apreciar que en el año 2021 este sube casi 20 puntos podemos inferir que es un efecto post-pandemia , ya que durante durante dicho año las restricciones viales estuvieron presentes.

```python
# Convertir la columna de fecha a tipo datetime
dfh1['FECHA'] = pd.to_datetime(dfh1['FECHA'])

# Crear una nueva columna para el año
dfh1['AÑO'] = dfh1['FECHA'].dt.year

# Filtrar los datos solo para las víctimas en motocicleta
victimas_moto = dfh1[dfh1['VICTIMA'] == 'PEATON']

# Calcular el número total de víctimas en motocicleta por año
victimas_moto_por_año = victimas_moto.groupby('AÑO').size()

# Crear el gráfico de tendencia de víctimas en motocicleta a lo largo de los años
plt.figure(figsize=(10, 6))
plt.plot(victimas_moto_por_año.index, victimas_moto_por_año.values, marker='o')
plt.title('Tendencia de víctimas tipo Peaton a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Total de víctimas Peatones')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
